```python
–ö–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—ã–µ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏

–ö -—Ç–∞—è –±–∞–Ω–∫–∞
–£ –ù–∏–∫–∏—Ç—ã –µ—Å—Ç—å ùëõ –±–∞–Ω–æ–∫ –≥–∞–∑–∏—Ä–æ–≤–∫–∏, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –∏–º–µ–µ—Ç —Å–≤–æ–π –æ–±—ä—ë–º.

–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –ø–æ–ª–µ–∑–Ω–µ–µ –≤—Å–µ–≥–æ –ø–∏—Ç—å –≥–∞–∑–∏—Ä–æ–≤–∫—É, –Ω–∞—á–∏–Ω–∞—è –æ—Ç –±–∞–Ω–æ–∫ –±–æ–ª—å—à–µ–≥–æ –æ–±—ä–µ–º–∞ –∏ –ø–µ—Ä–µ—Ö–æ–¥—è –∫ –≤—Å—ë –º–µ–Ω—å—à–∏–º.

–ü–æ–º–æ–≥–∏—Ç–µ –ù–∏–∫–∏—Ç–µ –Ω–∞–π—Ç–∏ ùëò-—é –ø–æ –ø–æ–ª–µ–∑–Ω–æ—Å—Ç–∏ –±–∞–Ω–∫—É. –í—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω—ã –¥–≤–∞ —á–∏—Å–ª–∞ 1‚â§ùëõ‚â§104 –∏ 1‚â§ùëò‚â§100. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ ùëò‚â§ùëõ.

–î–∞–ª–µ–µ –≤ —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω—ã ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª 0‚â§ùëéùëñ‚â§2‚ãÖ109 ‚Äî –æ–±—ä—ë–º—ã –±–∞–Ω–æ–∫.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–í—ã–≤–µ—Å—Ç–∏ –æ–±—ä—ë–º ùëò-–æ–π –≤ –ø–æ—Ä—è–¥–∫–µ –ø–æ–ª–µ–∑–Ω–æ—Å—Ç–∏ –±–∞–Ω–∫–∏.

–ü—Ä–∏–º–µ—Ä—ã
–í–≤–æ–¥:
5 5
1 7 2 3 2
–í—ã–≤–æ–¥:
1
```

n, k = map(int, input().split())
mas = list(map(int, input().split()))

def selection(mas, n, k):
  for i in range(n):
    idle = i
    for j in range (i, n):
        if mas[idle] < mas[j]:
          idle = j
    mas[i], mas[idle] = mas[idle], mas[i]
    if (i == k-1):
         return mas[:i+1]
print(selection(mas, n, k)[k-1])

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–º–µ–Ω–æ–≤

–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —Å–∫–æ–ª—å–∫–æ –æ–±–º–µ–Ω–æ–≤ —Å–¥–µ–ª–∞–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º –ø—É–∑—ã—Ä—å–∫–æ–≤–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ù–∞ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –¥–∞–Ω–æ —á–∏—Å–ª–æ ùëÅ(1‚â§ùëÅ‚â§1000) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –º–∞—Å—Å–∏–≤–µ. –ù–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî —Å–∞–º –º–∞—Å—Å–∏–≤. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –Ω–µ –ø—Ä–µ–≤—ã—à–∞—é—Ç –ø–æ –º–æ–¥—É–ª—é 109.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–º–µ–Ω–æ–≤ –ø—É–∑—ã—Ä—å–∫–æ–≤–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.

–í–≤–æ–¥
5
1 2 3 4 5 
–í—ã–≤–æ–¥
0
–í–≤–æ–¥
5
5 4 3 2 1 
–í—ã–≤–æ–¥
10

```python
n = int(input())
mas = list(map(int, input().split()))
count = 0
cc = 0
n = len(mas)
unsorted = True
while n > 1 and unsorted:
    unsorted = False
    for i in range(n-1):
        if mas[i] > mas[i+1]:
            mas[i], mas[i+1] = mas[i+1], mas[i]
            count += 1
            unsorted = True
    cc += 1
print(count)
```

–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏

–ë–∏–±–ª–∏–æ—Ç–µ—á–Ω—ã–π –º–µ—Ç–æ–¥
–ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ —Ä–∞–±–æ—Ç—É –º–µ—Ç–æ–¥–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—Å—Ç–∞–≤–∫–∞–º–∏ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é. –î–ª—è —ç—Ç–æ–≥–æ –≤—ã–≤–µ–¥–∏—Ç–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –¥–∞–Ω–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ –ø–æ—Å–ª–µ –∫–∞–∂–¥–æ–π –≤—Å—Ç–∞–≤–∫–∏ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã—Ö —Å—Ç—Ä–æ–∫–∞—Ö. –ï—Å–ª–∏ –ø–æ—Å–ª–µ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–∏—è –æ—á–µ—Ä–µ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Å—Å–∏–≤ –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è, —Ç–æ –µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ. –ï—Å–ª–∏ –º–∞—Å—Å–∏–≤ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –Ω–µ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∏—á–µ–≥–æ.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ù–∞ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –¥–∞–Ω–æ —á–∏—Å–ª–æ ùëÅ (1‚â§ùëÅ‚â§100) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –º–∞—Å—Å–∏–≤–µ. –ù–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω —Å–∞–º –º–∞—Å—Å–∏–≤: –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –Ω–µ –ø—Ä–µ–≤—ã—à–∞—é—â–∏—Ö 109.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–í—ã–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫–∏ (–ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤—Å—Ç–∞–≤–æ–∫) –ø–æ ùëÅ —á–∏—Å–µ–ª –∫–∞–∂–¥–∞—è.

```python
n = int(input())
mas = list(map(int, input().split()))

for i in range(1, n):
    tmp = mas[i]
    j = i -1
    while j >= 0 and mas[j] > tmp:
        mas[j+1] = mas[j]
        j -=1
    if mas[j+1] != tmp:
        mas[j+1] = tmp
        print(*mas)
```

–ù–∞ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –¥–∞–Ω–æ —á–∏—Å–ª–æ ùëÅ(1‚â§ùëÅ‚â§1000) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤. –ù–∞ –∫–∞–∂–¥–æ–π —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–µ –¥–∞–Ω—ã –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã–π –Ω–æ–º–µ—Ä –∏ –Ω–∞–±—Ä–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±–∞–ª–ª–æ–≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–≥–æ —É—á–∞—Å—Ç–Ω–∏–∫–∞. –í—Å–µ —á–∏—Å–ª–∞ –≤–æ –≤—Ö–æ–¥–Ω–æ–º —Ñ–∞–π–ª–µ –Ω–µ –ø—Ä–µ–≤—ã—à–∞—é—Ç 105.

–í—ã–≤–µ–¥–∏—Ç–µ –∏—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è –±–∞–ª–ª–æ–≤. –ï—Å–ª–∏ —É –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –±–∞–ª–ª—ã, —Ç–æ –∏—Ö –º–µ–∂–¥—É —Å–æ–±–æ–π –Ω—É–∂–Ω–æ –≤—ã–≤–æ–¥–∏—Ç—å –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã—Ö –Ω–æ–º–µ—Ä–æ–≤.

```python
n = int(input())
mas = []
for i in range(n):
    mas.append(list(map(int, input().split())))

for i in range(n):
    for j in range(n-1):
        if mas[j][1] < mas[j + 1][1]:
            mas[j], mas[j+1] = mas[j+1], mas[j]
        elif mas[j][1] == mas[j + 1][1]:
            if mas[j][0] > mas[j + 1][0]:
                mas[j], mas[j+1] = mas[j+1], mas[j]

for i in range(n):
    print(mas[i][0], ' ', mas[i][1])
```
